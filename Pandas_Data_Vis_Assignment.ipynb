**Create an area plot of all the columns for just the rows up to 20.**

```python
# Code goes here!
```

```python
df.loc[0:20].plot.area(alpha=0.37, figsize=(5,3))
```

```python
df.loc[10:20].plot.area(alpha=0.37, figsize=(5,3))
```

**Try moving the legend in the plot above and display it outside the plot, shown below** [Hint](http://stackoverflow.com/questions/23556153/how-to-put-legend-outside-the-plot-with-pandas) also [gca](https://stackoverflow.com/questions/27221009/fig-gca-vs-fig-add-subplot).....*If `f` is a variable holding a figure, `f.gca()` returns the axes associated with the figure.*

```python
# Code goes here!

f = plt.figure()

df.loc[:20].plot(kind='area', ax=f.gca(), alpha=0.37, figsize=(5,3))

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
f = plt.figure()

df.loc[10:20].plot(kind='area', ax=f.gca(), alpha=0.37, figsize=(5,3))

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
